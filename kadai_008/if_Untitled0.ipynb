```python
import random

# 変数varは整数
var = random.randint(1, 15)

print(var)

# 変数varが3の倍数と5の倍数に該当する場合はFizzBuzz (先にチェック)
if 0 == var % 3 and 0 == var % 5:
  print("FizzBuzz")
# 変数varが3の倍数はFizz
elif 0 == var % 3:
  print("Fizz")
# 変数varが5の倍数はBuzze
elif 0 == var % 5:
  print("Buzze")
# 該当なし
else:
  print("該当なし")
```
